```python
import data_analysis_functions as daf
import pandas as pd


df = pd.DataFrame()

for type_val in range(2, 8):
    new_df = daf.files_into_df(quantity=20, type=type_val)
    new_df = new_df.transpose()
    new_df['type'] = type_val 
    df = pd.concat([df, new_df.abs()], axis=0)

df = df.sample(frac=1).reset_index(drop=True)
    
```

## ML

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# Assuming 'type' is the target column
target_column = 'type'

# Assuming the rest of the columns are features
features = df.drop(target_column, axis=1)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, df[target_column], test_size=0.2, random_state=42)

lst = []
lst1 = []
# Initialize the KNN classifier (you can adjust the number of neighbors, 'n_neighbors', as needed)
for i in range(1, 21):
    knn_classifier = KNeighborsClassifier(n_neighbors=i)


# Train the classifier on the training data
    knn_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = knn_classifier.predict(X_test)
    y_pred1 = knn_classifier.predict(X_train)
    # Evaluate the accuracy of the model
    accuracy = accuracy_score(y_test, y_pred)
    lst.append(accuracy)
    accuracy1 = accuracy_score(y_train, y_pred1)
    lst1.append(accuracy1)
    # print(f'Accuracy: {accuracy:.2f}')

plt.figure(figsize=(15,10))
plt.grid(color='orange', linestyle='-', linewidth=1)
plt.scatter(x=np.arange(1, 21) ,y=lst)
plt.scatter(x=np.arange(1, 21) ,y=lst1, color='red')
# plt.scatter(lst1, color='red')


```

```python
difference = [abs(lst1[i] - lst[i]) for i in range(20)]
print(min(difference))
print(difference.index(min(difference)))
plt.figure(figsize=(15,10))
plt.grid(color='orange', linestyle='-', linewidth=1)
plt.scatter(x=np.arange(1, 21) ,y=difference)
```

```python
models = {}
history = {}

# Logistic Regression
from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression(max_iter=3000)

# Support Vector Machines
from sklearn.svm import LinearSVC
models['Support Vector Machines'] = LinearSVC()

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
models['K-Nearest Neighbor'] = KNeighborsClassifier()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy, precision, recall = {}, {}, {}

for key in models.keys():
    
    # Fit the classifier
    models[key].fit(X_train, y_train)
    
    # Make predictions
    predictions = models[key].predict(X_test)
    
    # Calculate metrics
    accuracy[key] = accuracy_score(predictions, y_test)
    precision[key] = precision_score(predictions, y_test,average='micro')
    recall[key] = recall_score(predictions, y_test,average='micro')
```

```python
df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])
df_model['Accuracy'] = accuracy.values()
df_model['Precision'] = precision.values()
df_model['Recall'] = recall.values()

df_model
```

```python
ax = df_model.plot.barh()
ax.legend(
    ncol=len(models.keys()), 
    bbox_to_anchor=(0, 1), 
    loc='lower left', 
    prop={'size': 14}
)
plt.tight_layout()
```
